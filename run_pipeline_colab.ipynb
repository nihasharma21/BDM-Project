#RESULTS

This project successfully developed and executed an end-to-end quantitative portfolio optimization pipeline using real market data. The final results tell us how statistical return analysis, risk measurement, and mathematical optimization can be combined to construct a diversified, risk-controlled investment portfolio.                                                                               
1. Market Data & Return Behavior:
Daily price data for 15 major equities across technology, communications, and real estate sectors was collected over the 2022–2024 time horizon. From this:
Daily simple and log returns were computed for all assets.
These were aggregated into monthly returns (23 periods × 15 assets).
The time-series plots show:
Significant volatility in growth stocks (e.g., META, NFLX, PLTR).
More stable return behavior in real estate and telecom equities (e.g., VTR, DLR, T).                                                                                
✅ Key Result:
The dataset captures both high-growth/high-risk assets and low-volatility defensive assets, enabling meaningful diversification through optimization.                                                                               
2. Correlation & Risk Structure Results:
A full correlation and covariance matrix (15 × 15) was constructed from monthly returns.
The correlation heatmap shows:
Strong clustering among technology stocks (AAPL, MSFT, ADBE, META, NFLX).
Lower correlations between real estate investment trusts (DLR, VTR, ESS, SBAC) and tech assets.                                                                               
✅ Key Result:
The portfolio contains true diversification opportunities, as several asset groups exhibit low or even near-zero correlation with one another — a necessary condition for effective risk reduction.
    
3. Portfolio Optimization Model Results:
A mean-variance portfolio optimization model was implemented using Pyomo and the IPOPT solver. The model included:
15 decision variables (one per asset).
Covariance-based risk minimization as the objective.
A maximum allocation constraint of 20% per asset to prevent concentration.
The solver successfully generated:
A feasible optimized portfolio allocation.
A solution that:
Caps individual stock risk
Forces cross-sector diversification
Balances growth stocks with defensive real estate and telecom equities                                                                               
✅ Key Result:
The optimized portfolio minimizes overall portfolio variance subject to diversification constraints, demonstrating mathematically sound, implementable asset allocation.
        
4. Overall Performance Interpretation:
From a systems and modeling perspective, this project shows that:
Financial return behavior is highly volatile and non-uniform across sectors.

Mathematical optimization can translate:
Market risk
Statistical relationships
